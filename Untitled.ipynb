```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
metal=['Au','Ag','Cu']
rotamero=['Ph','Pc','Pn']
colores={'Au':'gold','Ag':'grey','Cu':'brown'}
energias={'Au':{'Pc':0.54,'Ph':0.0,'Pn':0.88},
          'Ag':{'Pc':0.43,'Ph':0.0,'Pn':0.88},
          'Cu':{'Pc':0.42,'Ph':0.0,'Pn':2.18}}
alturas={'Au':{'Pc':0.425,'Ph':0.23,'Pn':0.575},
          'Ag':{'Pc':0.4,'Ph':0.23,'Pn':0.6},
          'Cu':{'Pc':0.35,'Ph':0.23,'Pn':0.675}}
pos_etiqueta={'Au':{'Pc':{'x':0.475,'y':0.375},'Ph':{'x':0.25,'y':0.175},'Pn':{'x':0.725,'y':0.5}},
              'Ag':{'Pc':{'x':0.475,'y':0.345},'Ph':{'x':0.25,'y':0.175},'Pn':{'x':0.725,'y':0.5}},
              'Cu':{'Pc':{'x':0.475,'y':0.25},'Ph':{'x':0.25,'y':0.1575},'Pn':{'x':0.725,'y':0.65}}}

pos_rot={'Au':{'Pc':{'x':0.5,'y':0.325},'Ph':{'x':0.27,'y':0.125},'Pn':{'x':0.75,'y':0.45}},
              'Ag':{'Pc':{'x':0.5,'y':0.275},'Ph':{'x':0.27,'y':0.12},'Pn':{'x':0.75,'y':0.45}},
              'Cu':{'Pc':{'x':0.5,'y':0.2},'Ph':{'x':0.27,'y':0.1},'Pn':{'x':0.75,'y':0.6}}}

for i in metal:
    fig, ax = plt.subplots(figsize=(15,10))
    posicion=0
    maximo=np.array([energias['Au'][j] for j in energias['Au']]).max()
    plt.ylim(-0.1,energias[i]['Pn']+(maximo*1.2))
    #plt.xticks([])
    #plt.tick_params(top=False, bottom=False, left=True, right=False, labelleft=True, labelbottom=False) 
    for j in rotamero:
        etiqueta="+"+str(energias[i][j])+" eV"
        plt.figtext(pos_rot[i][j]['x'],pos_rot[i][j]['y'],  'P$_{'+j[1].upper()+'}$', fontsize=20) 
        plt.figtext(pos_etiqueta[i][j]['x'],pos_etiqueta[i][j]['y'],  etiqueta, fontsize=15)
    for j in rotamero:
        altura=np.full((100,),energias[i][j])
        base=np.linspace(posicion,(posicion+1),100)
        ax.plot(base,altura,'-',linewidth=10,color=colores[i])
        plt.axis('off')
        newax = fig.add_axes([posicion/4+0.1,alturas[i][j],0.35, 0.35], anchor='C', zorder=1)
        figura='C-'+i+'-'+j+'.png'
        im = plt.imread(figura)
        newax.imshow(im)
        newax.axis('off')
        posicion=posicion+1
    fig.savefig('Comparacion'+i+'.png')
```
